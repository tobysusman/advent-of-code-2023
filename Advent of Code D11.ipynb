Import File

```python
with open("D11 Input.txt", "r") as file:
    file_contents = file.read()
# print(file_contents)
ls = file_contents.split("\n")
ls[:10]
```

```python
original_grid = ls

updated_grid = []
for original_line in original_grid:
    if list(set(original_line)) == ['.']:
        updated_grid.append(original_line)
    updated_grid.append(original_line)

empty_columns = []
column_chart = []

for i in range(len(updated_grid[0])):
    column = []
    for line in updated_grid:
        column.append(line[i])
    column_chart.append(column)

updated_grid = []
for column in column_chart:
    if list(set(column)) == ['.']:
        updated_grid.append(column)
    updated_grid.append(column)
print(updated_grid)
```

Find the location of all the galaxies:

```python
galaxy_spots = []
line_counter = 0
for line in updated_grid:
    entry_counter = 0
    for entry in line:
        if entry == '#':
            galaxy_spots.append([line_counter, entry_counter])
        entry_counter += 1
    line_counter += 1

galaxy_spots[:10]
```

Now to compare all the distances.

```python
import itertools
```

```python
galaxy_pairings = []
for galaxy_pair in itertools.combinations(galaxy_spots, 2):
    # Measure the distance between the two galaxies
    galaxy_1 = galaxy_pair[0]
    galaxy_2 = galaxy_pair[1]
    distance = abs(galaxy_1[0] - galaxy_2[0]) + abs(galaxy_1[1] - galaxy_2[1])
    galaxy_pairings.append([galaxy_1, galaxy_2, distance])
galaxy_pairings[:10]
```

Get the sum of them

```python
sum = 0
for pairing in galaxy_pairings:
    sum += pairing[2]
sum
```

```python
part_2_grid = ls
part_2_grid

blank_columns = []
for j in range(len(part_2_grid[0])):
    column = []
    for i in range(len(part_2_grid)):
        column.append(part_2_grid[i][j])
    if list(set(column)) == ['.']:
        blank_columns.append(j)
blank_columns  
```

```python
blank_rows = []
line_counter = 0
for line in part_2_grid:
    if list(set(line)) == ['.']:
        blank_rows.append(line_counter)
    line_counter += 1
# Runtime = 
blank_rows
```

Now do the same checks on pairs!

```python
part_two_galaxy_spots = []
line_counter = 0
for line in part_2_grid:
    entry_counter = 0
    for entry in line:
        if entry == '#':
            part_two_galaxy_spots.append([line_counter, entry_counter])
        entry_counter += 1
    line_counter += 1

part_two_galaxy_spots[:10]
```

```python
def blanks_crossed(galaxy_1, galaxy_2, blank_rows = blank_rows, blank_columns = blank_columns, multiplication_factor = 1000000):
    number_of_crossings = 0
    for blank_row in blank_rows:
        if blank_row > galaxy_1[0] and blank_row < galaxy_2[0] or blank_row < galaxy_1[0] and blank_row > galaxy_2[0]:
            number_of_crossings += 1
    for blank_column in blank_columns:
        if blank_column > galaxy_1[1] and blank_column < galaxy_2[1] or blank_column < galaxy_1[1] and blank_column > galaxy_2[1]:
            number_of_crossings += 1
    return number_of_crossings * (multiplication_factor - 1)
```

```python
part_two_galaxy_pairings = []
for galaxy_pair in itertools.combinations(part_two_galaxy_spots, 2):
    # Measure the distance between the two galaxies
    galaxy_1 = galaxy_pair[0]
    galaxy_2 = galaxy_pair[1]
    distance = abs(galaxy_1[0] - galaxy_2[0]) + abs(galaxy_1[1] - galaxy_2[1]) + blanks_crossed(galaxy_1, galaxy_2)
    part_two_galaxy_pairings.append([galaxy_1, galaxy_2, distance])
part_two_galaxy_pairings[:25]
```

```python
sum = 0
for pairing in part_two_galaxy_pairings:
    sum += pairing[2]
sum
```
