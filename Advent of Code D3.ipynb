## Advent of Code Day 3

```python
import pandas as pd
import re
```

Import File

```python
with open("D3 Input.txt", "r") as file:
    file_contents = file.read()
print(file_contents)
ls = file_contents.split("\n")
ls
```

We have a grid of 140x140.  I think the first step is to basically identify the _grid position_ of each symbol.  Then for each grid position of each symbol we can then cycle through all the numbers to identify if any of them are within 1, if so then put that number on the list.

```python
symbol_locations = []

line_counter = 0
for line in ls:
    word_counter = 0
    for word in line:
        # if word in ('*', '+', '-', '/', '$', '@', '#', '=', '%'):
        if word not in ('1', '2', '3', '4', '5', '6', '7', '8', '9', '0', '.'):
            symbol_locations.append([word, [line_counter, word_counter]])
        word_counter += 1
    line_counter += 1
symbol_locations
```

Okay - now we need to identify all strings of numbers:

```python
number_locations = []
line_counter = 0
for line in ls:
    words_in_line = re.finditer(r'\d+', line)
    numbers = [[m.string[m.start(0):m.end(0)],[m.start(0), m.end(0), line_counter]] for m in words_in_line]
    number_locations.append([numbers])
    line_counter += 1
number_locations
```

Okay - now we need to cycle through each of these numbers, if any of them are within one of any of the symbol rows and columns then we add it i nour list to be added.

```python
scope_numbers = []
for line in number_locations:
    for entry in line[0]:
        number = entry[0]
        column_from = entry[1][0]
        column_to = entry[1][1]
        row = entry[1][2]
        for symbol in symbol_locations:
            symbol_name = symbol[0]
            symbol_row = symbol[1][0]
            symbol_column = symbol[1][1]
            if abs(symbol_row - row) <= 1:
                for i in range(column_from, column_to):
                    if abs(symbol_column - i) <= 1:
                        scope_numbers.append([number, row])
                        break
scope_numbers
            
```

Add up all these numbers to submit the answer.

```python
sum_counter = 0
for number in scope_numbers:
    sum_counter += int(number[0])
sum_counter
```

Part 2

Okay, this is now fiddly.  I need to filter the symbols by only those that are `*`, then identify all connecting numbers.  If a number connects then it is added to a list. If there are more than 1 for that symbol then multiply them together.

```python
scope_things = []
symbol_counter = 0
for symbol in symbol_locations:
    if symbol[0] == '*':
        symbol_row = symbol[1][0]
        symbol_column = symbol[1][1]
        for number in number_locations:
            for entry in number[0]:
                number = entry[0]
                column_from = entry[1][0]
                column_to = entry[1][1]
                for i in range(column_from, column_to):
                    if abs(symbol_column - i) <= 1 and abs(symbol_row - entry[1][2]) <= 1:
                        scope_things.append([number, symbol_counter])
                        break
    symbol_counter += 1
print(scope_things)
```

```python
df = pd.DataFrame(scope_things, columns = ['Number', 'Symbol'])
df = df.groupby('Symbol')['Number'].apply(list).reset_index()
df['Length'] = df['Number'].apply(len)
df = df[df['Length'] >= 2]
product_numbers = df['Number'].tolist()

total_counter = 0
for product_number in product_numbers:
    total_counter += int(product_number[0]) * int(product_number[1])
total_counter
```
