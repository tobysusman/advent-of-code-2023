Define a function to identify every instance where a word appears.

```python
def findall(f, s):
    l = []
    i = -1
    while True:
        i = s.find(f, i+1)
        if i == -1:
            return l
        l.append(s.find(f, i))
```

```python
def findanymatches(numbers, text):
    matches = []
    for number in numbers:
        if findall(number, text) != []:
            matches.append(findall(number, text))
    return matches
```

```python
dict = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9}
numbers = dict.keys()

item = 'xtwone3four'

number_to_replace = ''
ls_new = []


for item in ls:
    item_to_update = item
    number_potential_matches = len(findanymatches(numbers, item_to_update))
    while number_potential_matches > 0:
        earliest_spot = float('inf')
        for number in numbers:
            if len(findall(number, item_to_update)) > 0:
                value = findall(number, item_to_update)[0]
                if value == -1:
                    pass
                elif value < earliest_spot:
                    earliest_spot = value
                    number_to_replace = number
            else:
                pass
        item_to_update = item_to_update.replace(number_to_replace, str(dict[number_to_replace]))
        number_potential_matches = len(findanymatches(numbers, item_to_update))
    ls_new.append(item_to_update)
ls_new
```
