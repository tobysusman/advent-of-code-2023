## Advent of Code Day 2

```python
import pandas as pd
import re
```

Import File

```python
with open("D2 Input.txt", "r") as file:
    file_contents = file.read()
print(file_contents)
ls = file_contents.split("\n")
ls
```

Now let's set things up a bit better.

```python
structured_ls = []
for i in ls:
    structured_ls.append(re.split(": ", i))

split_by_grab = []

def cube_counter(cubes_ls):
    if len(cubes_ls) == 0:
        return 0
    else:
        return int(cubes_ls[0])

for i in structured_ls:
    grab_counter = 1
    for grab in re.split("; ", i[1]):
        red_cubes = re.findall(r"(\d+) red", grab)
        blue_cubes = re.findall(r"(\d+) blue", grab)
        green_cubes = re.findall(r"(\d+) green", grab)
        split_by_grab.append([i[0], grab_counter, grab, cube_counter(red_cubes), cube_counter(blue_cubes), cube_counter(green_cubes)])
        grab_counter += 1
split_by_grab
```

Now put them all in a dataframe.

```python
df = pd.DataFrame(split_by_grab, columns = ['Game Number', 'Grab Number', 'String', 'Red Cubes', 'Blue Cubes', 'Green Cubes'])
df
```

Filter this down to identify any entries that are impossible and sum the unique game numbers.

```python
impossible_games_df = df[(df['Red Cubes'] > 12) | (df['Blue Cubes'] > 14) | (df['Green Cubes'] > 13)]
impossible_games_list = impossible_games_df['Game Number'].unique()
all_games_list = df['Game Number'].unique()

possible_games_list = [x for x in all_games_list if x not in impossible_games_list]

counter = 0
for i in possible_games_list:
    counter += int(re.findall(r"Game (\d+)", i)[0])
counter
```

Part 2

We need to find the multiplication of the highest number of each type of cube in each game. Then sum together for all games.  I think my df is structured pretty nicely for this.

```python
def min_cube_colour(df, game, colour_cubes):
    game_specific_df = df[df['Game Number'] == game]
    return game_specific_df[colour_cubes].max()

min_cube_colour(df, "Game 1", "Red Cubes")

power_counter = 0
for game in all_games_list:
    power_counter += min_cube_colour(df, game, "Red Cubes") * min_cube_colour(df, game, "Blue Cubes") * min_cube_colour(df, game, "Green Cubes")

power_counter
```
